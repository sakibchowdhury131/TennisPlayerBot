```python
import pybullet as p
import time

# Connect to the physics server
physicsClient = p.connect(p.GUI)  # or p.DIRECT for non-graphical version
p.setGravity(0, 0, -9.8)  # Set gravity

# Load a sphere (ball)
ball_id = p.createCollisionShape(p.GEOM_SPHERE, radius=0.5)
ball_visual_id = p.createVisualShape(p.GEOM_SPHERE, radius=0.5, rgbaColor=[1, 0, 0, 1])
ball_body_id = p.createMultiBody(baseMass=1.0, baseCollisionShapeIndex=ball_id, baseVisualShapeIndex=ball_visual_id, basePosition=[0, 0, 5])

# Simulation parameters
time_step = 1.0 / 240.0  # Set the simulation time step
num_simulation_steps = 1000  # Number of simulation steps

# Bouncing parameters
force_magnitude = 10.0  # Magnitude of the force applied for bouncing

# Simulate the bouncing ball
for _ in range(num_simulation_steps):
    # Step simulation
    p.stepSimulation()
    time.sleep(time_step)

    # Check the position of the ball
    pos, _ = p.getBasePositionAndOrientation(ball_body_id)

    # If the ball is below a certain height, apply an upward force to simulate bouncing
    if pos[2] < 0.5:
        p.applyExternalForce(ball_body_id, -1, forceObj=[0, 0, force_magnitude], posObj=[0, 0, 0], flags=p.WORLD_FRAME)

# Disconnect from the physics server
p.disconnect()
```

```python
import pybullet as p
import time
import math

# Connect to the physics server
physicsClient = p.connect(p.GUI)  # or p.DIRECT for non-graphical version
p.setGravity(0, 0, -9.8)  # Set gravity

# Load a plane
plane_id = p.createCollisionShape(p.GEOM_PLANE)
plane_visual_id = -1  # Plane doesn't need a visual representation
plane_body_id = p.createMultiBody(baseMass=0, baseCollisionShapeIndex=plane_id, baseVisualShapeIndex=plane_visual_id, basePosition=[0, 0, 0])

# Load a strong surface (box) with a mass of 50 kg
surface_id = p.createCollisionShape(p.GEOM_BOX, halfExtents=[1, 1, 0.1])
surface_visual_id = -1  # No visual representation for now
surface_body_id = p.createMultiBody(baseMass=50.0, baseCollisionShapeIndex=surface_id, baseVisualShapeIndex=surface_visual_id, basePosition=[0, 0, 0.1])

# Load a sphere (projectile)
projectile_id = p.createCollisionShape(p.GEOM_SPHERE, radius=0.1)
projectile_visual_id = p.createVisualShape(p.GEOM_SPHERE, radius=0.1, rgbaColor=[1, 0, 0, 1])
projectile_body_id = p.createMultiBody(baseMass=1.0, baseCollisionShapeIndex=projectile_id, baseVisualShapeIndex=projectile_visual_id, basePosition=[0, 0, 0.5])

# Set the restitution coefficient for the projectile
p.changeDynamics(projectile_body_id, -1, restitution=0.8)  # Adjust the restitution as needed (e.g., 0.8 for a good bounce)
p.changeDynamics(surface_body_id, -1, restitution=0.8)  # Adjust the restitution as needed (e.g., 0.8 for a good bounce)

# Initial conditions for projectile motion
initial_speed = 20.0  # Initial speed of the projectile
launch_angle = 45.0  # Launch angle in degrees

# Convert launch angle to radians
launch_angle_rad = math.radians(launch_angle)

# Calculate initial velocity components
initial_velocity = [0, 0, initial_speed]

# Apply initial force to simulate projectile motion
p.applyExternalForce(projectile_body_id, -1, forceObj=initial_velocity, posObj=[0, 0, 0], flags=p.LINK_FRAME)

# Simulation parameters
time_step = 1.0 / 240.0  # Set the simulation time step
num_simulation_steps = 1000  # Number of simulation steps

# Simulate projectile motion
for _ in range(num_simulation_steps):
    # Step simulation with a smaller time step
    p.stepSimulation()

    # Check the position of the projectile
    pos, _ = p.getBasePositionAndOrientation(projectile_body_id)

    # Stop simulation if the projectile is below the floor
    if pos[2] < 0:
        break

    time.sleep(time_step)

# Disconnect from the physics server
p.disconnect()
```

# Create Environment

```python
import pybullet as p
import pybullet_data
import numpy as np
```

```python
class TableTennisEnv:
    def __init__(self, show_gui = False):
        self.step_count = 0
        self.show_gui = show_gui
        self._plane = None
        self._surface_body_id = None
        self._table_body_id = None
        self._table2_body_id = None
        self._robotic_arm = None
        self._ball = None
        self.state = self.init_state()

        # Throw the ball
        # self.throw_ball()
        print(self.state)


    # get states at any time
    def get_state(self): 
        position_ball, orientation_ball = p.getBasePositionAndOrientation(self._ball)
        position_bat, orientation_bat, _, _, _, _ = p.getLinkState(self._robotic_arm, 9) ## it is the position of the end effector
        state = {
            'ball': {
                'position': position_ball,
                'orientation': orientation_ball,
            },
            'bat': {
                'position': position_bat,
                'orientation': orientation_bat,
            },
        }

        return state


    def init_state(self):
        if self.show_gui:
            p.connect(p.GUI)
        else:
            p.connect(p.DIRECT)

        p.resetSimulation()
        p.setAdditionalSearchPath(pybullet_data.getDataPath())
        p.setGravity(0, 0, -9.81)

        # load the plane
        self._plane = p.loadURDF('plane.urdf')
        p.changeDynamics(self._plane, -1, lateralFriction = 0.5)

        #load the platform
        surface_id = p.createCollisionShape(p.GEOM_BOX, halfExtents=[5, 3, 0.1])
        surface_visual_id = -1  # No visual representation for now
        self._surface_body_id = p.createMultiBody(baseMass=300.0, baseCollisionShapeIndex=surface_id, baseVisualShapeIndex=surface_visual_id, basePosition=[0, 0, 0.1])
        p.changeDynamics(self._surface_body_id, -1, restitution=0.8, lateralFriction = 0.5)  # Adjust the restitution as needed (e.g., 0.8 for a good bounce)


        # load the table for the tennis
        table_id = p.createCollisionShape(p.GEOM_BOX, halfExtents=[1.37, 0.75, 0.5])
        table_visual_id = -1  # No visual representation for now
        self._table_body_id = p.createMultiBody(baseMass=200.0, baseCollisionShapeIndex=table_id, baseVisualShapeIndex=table_visual_id, basePosition=[0, 0, 0.7])
        p.changeDynamics(self._table_body_id, -1, restitution=0.8, lateralFriction = 0.5)  # Adjust the restitution as needed (e.g., 0.8 for a good bounce)


        # load the secondary table for the robot
        table2_id = p.createCollisionShape(p.GEOM_BOX, halfExtents=[0.5, 0.5, 0.5])
        table2_visual_id = -1  # No visual representation for now
        self._table2_body_id = p.createMultiBody(baseMass=50.0, baseCollisionShapeIndex=table2_id, baseVisualShapeIndex=table2_visual_id, basePosition=[2, 0, 0.7])
        p.changeDynamics(self._table2_body_id, -1, restitution=0.8, lateralFriction = 0.5)  # Adjust the restitution as needed (e.g., 0.8 for a good bounce)


        # load the robotic arm
        self._robotic_arm = p.loadURDF('combined_robot.urdf', basePosition=[1.7,0,1.25], baseOrientation=[0,0,0,1], globalScaling=1.0, useFixedBase=True)
        p.changeDynamics(self._robotic_arm, 9, restitution = 0.8)


        # load the tennis ball
        self._ball = p.loadURDF('tennis_ball2/urdf/tennis_ball2.urdf', basePosition = [-1, 0, 2])
        p.changeDynamics(self._ball, -1, restitution = 0.8)


        self.state = self.get_state()
        return self.state


    # reset the environment to initial config
    def reset(self):
        p.disconnect()
        self.state = self.init_state()
        self.step_count = 0


    def shutdown(self):
        p.disconnect()

    
    # throw ball function
    def throw_ball(self, force_vector = [3.05, 0.8, 2.4]):  # set the force[force_x, force_y, force_z] # unit in Newton
        p.applyExternalForce(self._ball, -1, forceObj=force_vector, posObj=[0, 0, 0], flags=p.LINK_FRAME)
        # self.state = self.get_state()


    def step(self, action): # action = [joint1, joint2, joint3, joint4, joint5, joint6] --> joint angles in radian
        self.step_count += 1
        joint_1 = action[0]
        joint_2 = action[1]
        joint_3 = action[2]
        joint_4 = action[3]
        joint_5 = action[4]
        joint_6 = action[5]
        end_effector = 1 # useless
        finger1_joint1 = 1
        finger1_joint2 = 1
        finger2_joint1 = 1
        finger2_joint2 = 1
        finger3_joint1 = 1
        finger3_joint2 = 1
    
        joint_angles_vector = [0.0, 0.0, joint_1, joint_2, joint_3, joint_4, joint_5, joint_6,
                              end_effector, finger1_joint1, finger1_joint2, 
                              finger2_joint1, finger2_joint2,
                              finger3_joint1, finger3_joint2,]
        
        p.setJointMotorControlArray(self._robotic_arm, range(15), 
                                    p.POSITION_CONTROL,
                                    targetPositions=joint_angles_vector)

        # step
        p.stepSimulation()

        # update the state
        self.state = self.get_state()

        if self.step_count > 50000: # setting done condition --> episode ends
            # set reward
            self.reset()
            reward = -1  # arbitrary reward currently, needs to be updated
            done = True
            return reward, done
        
        # if not done
        done = False
        reward = -1 # arbitrary reward

        return reward, done
```
